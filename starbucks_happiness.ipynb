# Clean Starbucks dataset

## Eliminate unnecessary columns

```python
test = starbucks[['Store Number','Brand', 'Store Name','Ownership Type','Street Address','City','State/Province','Country']]
test.head()
```

```python
pycountry.countries.lookup(test['Country'][0])
```

```python
df = test

list_alpha_2 = [i.alpha_2 for i in list(pycountry.countries)]
# list_alpha_3 = [i.alpha_3 for i in list(pycountry.countries)]    

inv_count = 0

def country_flag(df):
    if (len(df['Country'])==2 and df['Country'] in list_alpha_2):
        return pycountry.countries.get(alpha_2=df['Country']).name
    else:
        inv_count += 1
        return 'Invalid Code'

print(f"invalid a_2 count:{inv_count}")
df['country_2']=df.apply(country_flag, axis = 1)
```

```python
starbucks =df[['Brand','Store Number', 'Store Name','Ownership Type','Street Address','City','State/Province','Country','country_2']].copy()
starbucks=starbucks.rename(columns={"Store Number":"id",
                          "Brand":"brand",
                         "Store Name":"store_name",
                         "Ownership Type":"ownership_type",
                         "Street Address":"address",
                         "City":"city",
                         "State/Province":"state",
                        "country_2":"country"})
```

```python
starbucks=starbucks[['id','brand', 'store_name','ownership_type','address','city','state','country']]
starbucks.set_index('id')
starbucks=starbucks[starbucks['id']!='47370-257954']
```

```python
print(f"# unique countires: {len(starbucks['country'].unique())}")
starbucks['country'].value_counts()
starbucks.head()
```
